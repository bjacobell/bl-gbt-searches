```python
import numpy as np
import matplotlib.pyplot as plt

# Set global plotting parameters for a nicer appearance
plt.rcParams['figure.figsize'] = (10, 6)      # Set default figure size
plt.rcParams['axes.titlesize'] = 'large'      # Title font size
plt.rcParams['axes.labelsize'] = 'medium'     # Axes label font size
plt.rcParams['xtick.labelsize'] = 'small'     # X-axis tick label size
plt.rcParams['ytick.labelsize'] = 'small'     # Y-axis tick label size
plt.rcParams['legend.fontsize'] = 'medium'    # Legend font size
plt.rcParams['lines.linewidth'] = 2           # Default line width
plt.rcParams['lines.markersize'] = 8          # Default marker size
plt.rcParams['grid.color'] = 'gray'           # Grid color
plt.rcParams['grid.linestyle'] = '--'         # Grid line style
plt.rcParams['grid.linewidth'] = 0.5          # Grid line width
plt.rcParams['grid.alpha'] = 0.5              # Grid transparency

# Enable interactive mode for interactivity in Jupyter notebooks
plt.ion()

# Create a figure and axis object
fig, ax = plt.subplots()

# Example plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
ax.plot(x, y, marker='o', label='Example Plot')

# Set labels and title
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Example Plot')

# Add grid
ax.grid(True)

# Add legend
ax.legend()

# Show plot
plt.show()
```
